```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file
df = pd.read_excel('car.xlsx', sheet_name='Sheet1')

# Display the first few rows of the dataframe
df.head()


```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Fuel Type'], df['Total emissions (Kg CO2e per mile)'], color='skyblue')
plt.xlabel('Fuel Type')
plt.ylabel('Total emissions (Kg CO2e per mile)')
plt.title('Total Emissions by Fuel Type')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Read the Excel file
df = pd.read_excel('car.xlsx', sheet_name='Sheet2')

# Display the first few rows of the dataframe
df
```

```python
miles_df = df.loc[df['Unit'] == 'miles']

# Display the filtered DataFrame
miles_df.head()

km_df = df.loc[df['Unit'] == 'km']

# Display the filtered DataFrame
km_df.head()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(miles_df['Type'], miles_df['kg CO2e Diesel'], marker='o', label='Miles')
plt.plot(km_df['Type'], km_df['kg CO2e Diesel'], marker='x', label='Km')

plt.title('CO2 Emissions for Different Types')
plt.xlabel('Type')
plt.ylabel('Kg CO2e')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Read the Excel file
df = pd.read_excel('car.xlsx', sheet_name='Sheet3')

# Display the first few rows of the dataframe
df.head()
```

```python
import matplotlib.pyplot as plt

# Data
labels = df['Homeworking Activity']
values = df['kg CO2e']

# Plotting
plt.figure(figsize=(8, 6))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('CO2 Emissions for Homeworking Activity per FTE Working Hour')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
